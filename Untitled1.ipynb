```python
from transformers import pipeline
generator = pipeline("text-generation",model="distilgpt2")
word=input("enter a starting word or phrase :").strip()
if not word:
  print("error : input cannot be empty. please restart and provide a valid input")
else:
    try:
      results = generator(word,max_length=30,num_return_sequences=2)
      print("\n generated texts:")
      for i,result in enumerate(results):
        print(f"\n option {i+1}:\n{result['generated_text']}")
    except Exception as e:
      print(f"an error occured :{e}")
```
